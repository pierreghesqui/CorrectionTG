### Exercice 1. Calculs de quantité de matière (sans calculatrice)

1.  **Quantité de matière d'ions oxonium ($\text{H}_3\text{O}^+$)**
    * Formule utilisée : $n = C \times V$
    * $C = 1,5 \text{ mmol}\cdot\text{L}^{-1}$
    * $V = 20 \text{ mL} = 0,020 \text{ L}$
    * $n(\text{H}_3\text{O}^+) = (1,5 \text{ mmol}\cdot\text{L}^{-1}) \times (0,020 \text{ L}) = 0,03 \text{ mmol}$

2.  **Quantité de matière de paracétamol**
    * Formule utilisée : $n = \frac{m}{M}$
    * $m = 500 \text{ mg} = 0,500 \text{ g}$
    * $M = 150 \text{ g}\cdot\text{mol}^{-1}$

    * Formule utilisée : $n = \frac{V}{V_m}$
    * $V = 1,0 \text{ L}$
    * $V_m = 24 \text{ L}\cdot\text{mol}^{-1}$
    * $n(\text{CO}_2) = \frac{1,0 \text{ L}}{24 \text{ L}\cdot\text{mol}^{-1}} \approx 0,0417 \text{ mol} = 41,7 \text{ mmol}$

4.  **Quantité de matière d'éthanol pur**

    * $M = 50 \text{ g}\cdot\text{mol}^{-1}$
    * $n(\text{éthanol}) = \frac{40 \text{ g}}{50 \text{ g}\cdot\text{mol}^{-1}} = 0,8 \text{ mol}$

### Exercice 5. (sans calculatrice)

Soit une transformation de la forme $5\text{A}+2\text{B}+\text{C}\rightleftharpoons2\text{E}+3\text{F}$.

**Rappel :** Rendement $\eta = \frac{n_{\text{produit obtenu}}}{n_{\text{produit théorique maximal}}} \times 100\%$. Pour une réaction non totale, $n_{\text{produit théorique maximal}}$ correspond à $n_{\text{produit au cas où } x=x_{max}}$.

1.  **Situation 1 : 12 µmol de E ont été fabriquées. 60 µmol de A a été introduites. A est le réactif limitant.**
    * Si A est le réactif limitant et la réaction était totale, $5x_{max} = n_i(\text{A}) = 60 \text{ µmol}$.
    * Donc $x_{max} = \frac{60}{5} = 12 \text{ µmol}$.
    * Quantité théorique maximale de E : $n_{\text{E,max}} = 2x_{max} = 2 \times 12 = 24 \text{ µmol}$.
    * Quantité de E obtenue : $n_{\text{E,obtenu}} = 12 \text{ µmol}$.
    * Rendement $\eta = \frac{12}{24} \times 100\% = 50\%$

2.  **Situation 2 : 21 mmol de F ont été fabriquées. 14 mmol de B ont été introduites. B est le réactif limitant.**
    * Si B est le réactif limitant et la réaction était totale, $2x_{max} = n_i(\text{B}) = 14 \text{ mmol}$.
    * Donc $x_{max} = \frac{14}{2} = 7 \text{ mmol}$.
    * Quantité théorique maximale de F : $n_{\text{F,max}} = 3x_{max} = 3 \times 7 = 21 \text{ mmol}$.
    * Quantité de F obtenue : $n_{\text{F,obtenu}} = 21 \text{ mmol}$.
    * Rendement $\eta = \frac{21}{21} \times 100\% = 100\%$

3.  **Situation 3 : 15 mmol de A ont été consommées. 20 mmol de A ont été introduites.**
    * Le réactif limitant n'est pas explicitement donné. Cependant, on peut calculer le rendement en réactif limitant si on considère que A est le réactif limitant ou en se basant sur la consommation du réactif.
    * La question demande le rendement, qui est défini par rapport au réactif limitant.
    * Si A est le réactif limitant, $n_{i}(\text{A}) = 20 \text{ mmol}$.
    * La quantité de A consommée est $15 \text{ mmol}$.
    * Si 15 mmol de A ont été consommés, et si la transformation était totale avec A comme limitant, alors 20 mmol auraient dû être consommés.
    * Rendement de consommation de A (si A est le limitant) : $\eta = \frac{15}{20} \times 100\% = 75\%$

4.  **Situation 4 : 12 mol de B ont été introduites. B est le réactif limitant. 3 mol de E ont été formées.**
    * Si B est le réactif limitant et la réaction était totale, $2x_{max} = n_i(\text{B}) = 12 \text{ mol}$.
    * Donc $x_{max} = \frac{12}{2} = 6 \text{ mol}$.
    * Quantité théorique maximale de E : $n_{\text{E,max}} = 2x_{max} = 2 \times 6 = 12 \text{ mol}$.
    * Quantité de E obtenue : $n_{\text{E,obtenu}} = 3 \text{ mol}$.
    * Rendement $\eta = \frac{3}{12} \times 100\% = 25\%$

Soit une transformation de la forme $5~\text{A}+2~\text{B}+\text{C}\rightarrow2~\text{E}+3~\text{F}$. On suppose que la transformation est totale.

**Rappel :** Si A et B sont introduits dans des proportions stœchiométriques, cela signifie que $\frac{n_i(\text{A})}{5} = \frac{n_i(\text{B})}{2}$. De plus, comme la transformation est totale, $x_f = x_{max}$.
